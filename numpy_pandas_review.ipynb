```python
#beşe beşlik bir matriks oluşturdum, içindeki en küçük değer 1 en büyük değer 349 olsun istedim
array3 = np.random.randint(1,350,size=(5,5))
array3
```

```python
#bir blok önce rastgele oluşturduğum matriksi bir dataframeye çevirdim index ve sütun isimlerini  ben belirledim
#dataframei exceldeki tablolar gibi düşünmeniz faydalı olacaktır
dataframe = pd.DataFrame(data = array3 , index = [1,2,3,4,5] , columns = ["a","b","c","d","e"])
dataframe
```

```python
#dataframeye yeni bir sütun ekliyorum, bilerek içine bir adet nan değer atıyorum 
dataframe["f"] = np.array([2,3,4,5,np.nan])
dataframe
```

```python
#isnull diyerek değerleri gezip null olan varmı bulduruyorum sum fonksiyonu ile kaç adet true döndüğünü buluyorum 
dataframe.isnull().sum()
```

```python
#null değeri doldurmak için ilk önce sütunların ortalamasını alıyorum ilk mean fonksiyonu buna yarıyor, sonra tüm sütunların
#ortalamasının ortalamasını alıyorum ikinci mean buna yarıyor
dataframe.mean().mean()
```

```python
#bir blok üstte bulduğum değeri fillna fonksiyonunu kullanarak null değerlerin yerine koyuyorum
dataframe.fillna(value= 170.125 , inplace = True)
```
